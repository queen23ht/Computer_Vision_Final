```python
import pandas as pd
import os
```

```python
ao_uni = "./archive/uniqlo_final/ao_uniqlo/"
quan_uni = "./archive/uniqlo_final/quan_uniqlo/"
ao_yame = "./archive/yame_final/ao_yame/"
quan_yame = "./archive/yame_final/quan_yame/"
ao_adidas = "./archive/adidas_final/ao_adidas/"
quan_adidas = "./archive/adidas_final/quan_adidas/"

adidas_csv = "./archive/adidas_final/adidas_final.csv"
uniqlo_csv = "./archive/uniqlo_final//uniqlo_final.csv"
yame_csv = "./archive/yame_final/yame_final.csv"
```

```python
ADIDAS_DF = pd.read_csv(adidas_csv)
list_adidas = os.listdir(ao_adidas) + os.listdir(quan_adidas)
at_first_shape = ADIDAS_DF.shape
```

```python
ADIDAS_DF.head()
```

```python
ADIDAS_DF['images'] = ADIDAS_DF['images'].apply(eval)
```

```python
# filter out the images that are not in the list

def check_in_images(x):
    for i in x:
        if i in list_adidas:
            return True
    return False

ADIDAS_DF = ADIDAS_DF[ADIDAS_DF['images'].apply(check_in_images)]
after_shape = ADIDAS_DF.shape
ADIDAS_DF
```

```python
print(f"Before: {at_first_shape}, After: {after_shape}")
```

```python
UNIQLO_DF = pd.read_csv(uniqlo_csv)
list_uniqlo = os.listdir(ao_uni) + os.listdir(quan_uni)
at_first_shape = UNIQLO_DF.shape

UNIQLO_DF.head()
```

```python
UNIQLO_DF['img_list'] = UNIQLO_DF['img_list'].apply(eval)
```

```python
# filter out the images that are not in the list
def uniqlo_list_check(x):
    for i in x:
        if i in list_uniqlo:
            return True
    return False

UNIQLO_DF = UNIQLO_DF[UNIQLO_DF['img_list'].apply(uniqlo_list_check)]
after_shape = UNIQLO_DF.shape
UNIQLO_DF.head()
```

```python
print(f"Before: {at_first_shape}, After: {after_shape}")
```

```python
YAME_DF = pd.read_csv(yame_csv)
list_yame = os.listdir(ao_yame) + os.listdir(quan_yame)

at_first_shape = YAME_DF.shape
YAME_DF.head()
```

```python
YAME_DF['img_list'] = YAME_DF['img_list'].apply(eval)
```

```python
def yame_check(x):
    for i in x:
        if i in list_yame:
            return True
    return False

YAME_DF = YAME_DF[YAME_DF['img_list'].apply(yame_check)]

after_shape = YAME_DF.shape
YAME_DF.head()
```

```python
print(f"Before: {at_first_shape}, After: {after_shape}")
```
