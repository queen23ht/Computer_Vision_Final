```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('adidas_final.csv')
```

```python
# Tóm tắt thống kê dữ liệu
df.describe(include='all')
```

```python
import plotly.express as px
top_5_colors = df['color'].value_counts().head(5)
fig = px.bar(top_5_colors, x=top_5_colors.index, y=top_5_colors.values, 
             labels={'x': 'Màu Sắc', 'y': 'Tần Suất'}, 
             title='Top 5 Màu Sắc Phổ Biến Nhất')

# Hiển thị biểu đồ
fig.show()

```

```python
top_10_titles = df['title'].value_counts().head(10)

# Vẽ biểu đồ bar
fig = px.bar(top_10_titles, x=top_10_titles.index, y=top_10_titles.values, 
             labels={'x': 'Tên Sản Phẩm', 'y': 'Tần Suất'}, 
             title='Top 10 Sản Phẩm Phổ Biến Nhất')

# Hiển thị biểu đồ
fig.show()
```

```python
# Thay thế NaN bằng chuỗi rỗng trước khi tính độ dài
df['decripsion'] = df['decripsion'].fillna('')

# Tính độ dài của mô tả
df['description_length'] = df['decripsion'].apply(len)

# Vẽ histogram cho độ dài mô tả
fig = px.histogram(df, x='description_length', nbins=30, 
                   labels={'description_length': 'Độ Dài Mô Tả'},
                   title='Phân Phối Độ Dài Mô Tả Sản Phẩm')

# Hiển thị biểu đồ
fig.show()

```

```python
# Tính số lượng ảnh cho mỗi sản phẩm
df['num_images'] = df['images'].apply(lambda x: len(eval(x)))

# Vẽ histogram cho số lượng ảnh
fig = px.histogram(df, x='num_images', nbins=10, 
                   labels={'num_images': 'Số Lượng Hình Ảnh'},
                   title='Phân Phối Số Lượng Hình Ảnh Cho Mỗi Sản Phẩm')

# Hiển thị biểu đồ
fig.show()
```

```python
# Kiểm tra số lượng giá trị thiếu cho mỗi cột
missing_values = df.isnull().sum()

# Vẽ biểu đồ bar cho các giá trị thiếu
fig = px.bar(missing_values[missing_values > 0], x=missing_values[missing_values > 0].index, y=missing_values[missing_values > 0].values,
             labels={'x': 'Tên Cột', 'y': 'Số Lượng Giá Trị Thiếu'},
             title='Số Lượng Giá Trị Thiếu Trong Các Cột')

# Hiển thị biểu đồ
fig.show()
```

```python
from wordcloud import WordCloud
import plotly.express as px
import matplotlib.pyplot as plt
from io import BytesIO

# Kết hợp tất cả mô tả thành một chuỗi
text = " ".join(df['decripsion'].dropna())

# Tạo WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Chuyển wordcloud thành hình ảnh
img = wordcloud.to_image()

# Hiển thị WordCloud bằng Plotly
buf = BytesIO()
img.save(buf, format="PNG")
buf.seek(0)

fig = px.imshow(plt.imread(buf))
fig.update_layout(title="Word Cloud Cho Cột 'Description'")
fig.update_xaxes(showticklabels=False).update_yaxes(showticklabels=False)
fig.show()

```

```python
df['word_count'] = df['decripsion'].apply(lambda x: len(str(x).split()))
fig = px.scatter(df, x='word_count', y='description_length', 
                 labels={'word_count': 'Số Lượng Từ', 'description_length': 'Độ Dài Mô Tả'},
                 title='Tương Quan Giữa Số Lượng Từ và Độ Dài Mô Tả')
fig.show()

```
