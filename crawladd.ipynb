```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Khởi tạo driver
driver = webdriver.Chrome()
driver.get("https://www.adidas.com.vn/vi/nam-ao_phong-quan_ao")

# Dừng 2 giây để trang web tải xong
time.sleep(2)

# Danh sách lưu trữ các liên kết hình ảnh
list_links_img = []

# Lặp qua các trang
start = 0
while True:
    # Lặp qua các ảnh trong trang hiện tại
    for i in range(1, 49):  # Mỗi trang có tối đa 48 sản phẩm
        try:
            img = driver.find_element(By.XPATH, f"/html/body/div[1]/div[2]/main/section[2]/article[{i}]/div/header/a/img[1]")
            link = img.get_attribute("src")
            list_links_img.append(link)
        except Exception as e:
            print(f"Error while scraping image {i}: {e}")
    
    # Cập nhật URL để chuyển sang trang tiếp theo
    start += 48  # Mỗi trang có 48 sản phẩm, tăng start mỗi lần
    next_page_url = f"https://www.adidas.com.vn/vi/nam-ao_phong-quan_ao?start={start}"
    
    # Kiểm tra nếu link trang tiếp theo còn hợp lệ
    driver.get(next_page_url)
    time.sleep(2)  # Đợi trang mới tải

    # Nếu không còn sản phẩm nào trên trang (ví dụ: không có 48 sản phẩm), dừng lại
    if len(driver.find_elements(By.XPATH, "/html/body/div[1]/div[2]/main/section[2]/article")) == 0:
        break

# In danh sách các liên kết hình ảnh đã thu thập được
print(list_links_img)
```

```python
import os
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Danh sách các link cần cào dữ liệu
urls = [
    "https://www.adidas.com.vn/vi/nam-ao_phong-quan_ao",
    "https://www.adidas.com.vn/vi/nam-ao_jersey-quan_ao",
    "https://www.adidas.com.vn/vi/nam-ao_khoac",
    "https://www.adidas.com.vn/vi/nam-bo_do_the_thao",
    "https://www.adidas.com.vn/vi/nam-hoodies",
    "https://www.adidas.com.vn/vi/nam-quan_short-quan_ao",
    "https://www.adidas.com.vn/vi/nam-sportswear-quan_ao"
]

# Khởi tạo driver
driver = webdriver.Chrome()

# Thư mục lưu trữ hình ảnh
base_dir = "crawled_images"
if not os.path.exists(base_dir):
    os.makedirs(base_dir)

# Lặp qua từng URL trong danh sách
for url in urls:
    driver.get(url)
    time.sleep(2)  # Đợi trang web tải xong
    
    # Lấy tên thể loại từ URL (ví dụ: "nam-ao_phong-quan_ao" -> "Áo Phông Quần Áo")
    category = url.split('/')[-1].replace('-', ' ').title()

    # Tạo thư mục cho thể loại nếu chưa có
    category_dir = os.path.join(base_dir, category)
    if not os.path.exists(category_dir):
        os.makedirs(category_dir)

    start = 0
    while True:
        # Lặp qua các ảnh trong trang hiện tại
        for i in range(1, 49):  # Mỗi trang có tối đa 48 sản phẩm
            try:
                img = driver.find_element(By.XPATH, f"/html/body/div[1]/div[2]/main/section[2]/article[{i}]/div/header/a/img[1]")
                img_url = img.get_attribute("src")
                
                # Tải ảnh về và lưu vào thư mục thể loại
                img_name = img_url.split("/")[-1]
                img_path = os.path.join(category_dir, img_name)
                
                # Tải ảnh bằng requests
                img_data = requests.get(img_url).content
                with open(img_path, 'wb') as f:
                    f.write(img_data)
                print(f"Downloaded: {img_name} to {category_dir}")
            except Exception as e:
                print(f"Error while downloading image {i} on {url}: {e}")
        
        # Cập nhật URL để chuyển sang trang tiếp theo
        start += 48  # Mỗi trang có 48 sản phẩm, tăng start mỗi lần
        next_page_url = f"{url}?start={start}"
        
        # Điều hướng đến trang tiếp theo
        driver.get(next_page_url)
        time.sleep(2)  # Đợi trang mới tải

        # Nếu không còn sản phẩm nào trên trang (ví dụ: không có 48 sản phẩm), dừng lại
        if len(driver.find_elements(By.XPATH, "/html/body/div[1]/div[2]/main/section[2]/article")) == 0:
            break

# Đóng driver khi xong
driver.quit()
```
